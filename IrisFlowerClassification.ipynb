```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn import tree
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
import numpy as np
```

```python
iris_data = pd.read_csv ('IRIS.csv')
```

```python
#dimensions (no. of rows & columns)
print(iris_data.shape)
# list of columns/features
print(iris_data.columns )
#peek some data
print(iris_data.head(10))
# statistical summary
print(iris_data.describe())
```

```python
# target variable
target =iris_data ['species']
# distribution of class labels or categories
print(pd.value_counts(target))
#alternative of finding class distribution
print(iris_data .groupby('species').size())
```

```python
seed = 4
train_data , test_data =  train_test_split(iris_data, test_size = 0.3, random_state = seed)
#shape of dataset
print('\n Shape of training data:', train_data.shape)
print( '\n Shape of testing data:', test_data.shape)
#class distribution of training data
print('\n Class distribution of training data:')
print(train_data.groupby('species').size())
#class distribution of testing data
print('\n Class distribution of testing data:')
print(test_data.groupby('species').size())
```

```python
train_data , test_data = train_test_split(iris_data, test_size=0.3, random_state=seed, stratify=target)
#shape of dataset
print('\n Shape of training data:', train_data.shape)
print( '\n Shape of testing data:', test_data.shape)
#class distribution of training data
print('\n Class distribution of training data:')
print(train_data.groupby('species').size())
#class distribution of testing data
print('\n Class distribution of testing data:')
print(test_data.groupby('species').size())
```

```python
#separate the independent and target variables from training data
train_x= train_data.drop(columns=['species'], axis=1)
train_y= train_data['species']

#separate the independent and target variables from test data test_x
test_x= test_data.drop(columns=['species'], axis=1)
test_y= test_data['species']
```

```python
#create a classifier object/ model
model = tree.DecisionTreeClassifier()
#trian the data with fit function
model.fit(train_x, train_y)
```

```python
#make prediction on train data
predictions_train= model.predict(train_x)
trainA=  accuracy_score(predictions_train, train_y)

#make prediction on test data
predictions_test= model.predict(test_x)
testA= accuracy_score(predictions_test, test_y)
trainingAccuracy=[]
testingAccuracy=[]
trainingAccuracy.append(trainA*100)
testingAccuracy.append(testA*100)
```

```python
# Create a Logistic Regression classifier object/model
model = LogisticRegression(max_iter=1000)

# Train the model using the fit function
model.fit(train_x, train_y)
```

```python
# make predictions on training data
predictions_train = model.predict(train_x)
trainA=accuracy_score(train_y, predictions_train)*100
trainingAccuracy.append(trainA)
# make predictions on test data
predictions_test = model.predict(test_x)
testA=accuracy_score(test_y, predictions_test)*100
testingAccuracy.append(testA)
```

```python
# Create a kNN classifier object/model
model = KNeighborsClassifier(n_neighbors=5)  # You can adjust the number of neighbors (k)

# Train the model using the fit function
model.fit(train_x, train_y)
```

```python
# make predictions on training data
predictions_train = model.predict(train_x)
trainA=accuracy_score(train_y, predictions_train)*100
trainingAccuracy.append(trainA)
# make predictions on test data
predictions_test = model.predict(test_x)
testA=accuracy_score(test_y, predictions_test)*100
testingAccuracy.append(testA)
```

```python
# Create an SVM classifier object/model
model = SVC()  # You can specify kernel, C, gamma, etc.

# Train the model using the fit function
model.fit(train_x, train_y)
```

```python
# make predictions on training data
predictions_train = model.predict(train_x)
trainA=accuracy_score(train_y, predictions_train)*100
trainingAccuracy.append(trainA)
# make predictions on test data
predictions_test = model.predict(test_x)
testA=accuracy_score(test_y, predictions_test)*100
testingAccuracy.append(testA)
```

```python
models = ['Decession Tree', 'Logistic Regression', 'kNN', 'Support Vector Machine']

df = pd.DataFrame({
    'Model': models,
    'Training Accuracy': trainingAccuracy,
    'Test Accuracy': testingAccuracy
})


print(df)
```
